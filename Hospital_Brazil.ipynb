```python
#-- creating a class to transpond the column CONVÊNIO SUS?
class CategoricalToDummyFeaturesTransform(BaseEstimator, TransformerMixin):
    """ Classe de Feature Transformer que transforma labels na representação Dummy.
        Esse imputer mantém a entrada X como um DataFrame em vez de transformar em numpy.array.
    """
    
    def __init__(self, categories):
        self.categories = [f"CONVÊNIO SUS?: {c}" for c in categories]
        
    def fit(self, X, y=None, **fit_params):
        return self
    
    def transform(self, X):        
        X_t = X.drop("CONVÊNIO SUS?", axis=1)
        
        dummy = pd.get_dummies(
            X["CONVÊNIO SUS?"], 
            prefix="CONVÊNIO SUS?",
            prefix_sep=": "
        ) 
        return X_t.join(dummy.loc[:, self.categories])
```

```python
#-- transponding CONVÊNIO SUS? column

""" Complete os espaços com ? """

categories = ["NÃO", "SIM"]

db_pib_gsus = CategoricalToDummyFeaturesTransform(categories).transform(db_pib_gsus)
```

```python
#-- printing the head
db_pib_gsus.head()
db_pib_gsus.shape
```

```python
#-- multiplying the column Count by CONVÊNIO SUS? NÃO and by CONVÊNIO SUS? SIM
db_pib_gsus['CONVÊNIO SUS?: NÃO'] = db_pib_gsus['Count'] * db_pib_gsus['CONVÊNIO SUS?: NÃO']
db_pib_gsus['CONVÊNIO SUS?: SIM'] = db_pib_gsus['Count'] * db_pib_gsus['CONVÊNIO SUS?: SIM']
```

```python
#-- importing KMeans and numpy
from sklearn.cluster import KMeans
import numpy as np

#-- defining number of clusters
n = 3
kmeans = KMeans(n_clusters=n)
```

```python
#-- droping string column
db_pib_gsus_merge_c = db_pib_gsus_merge.drop(['Concat'], axis=1)
```

```python
#-- fitting the model
kmeans.fit(db_pib_gsus_merge_c.values)
labels = kmeans.labels_
```

```python
#-- ploting graph with 3 clusters
plt.scatter(x=db_pib_gsus_merge_c.values[:,0],
            y=db_pib_gsus_merge_c.values[:,1],
            c=labels.astype(np.float),
            cmap='rainbow',
            edgecolor='k')
plt.title('Num clusters: {}'.format(n))
plt.show()
```

```python
#-- import lib to standardizing data
from sklearn.preprocessing import StandardScaler
```

```python
#-- plotting results
n_clusters = range(2,6)

X = db_pib_gsus_merge_c.values

# Dados normalizados
X_scaled = StandardScaler().fit_transform(X)

for n in n_clusters:
    estimator = KMeans(n_clusters=n, random_state=170)
    estimator.fit(X_scaled)
    labels = estimator.labels_
    plt.scatter(x=X_scaled[:,0],
                y=X_scaled[:,1],
                c=labels.astype(np.float),
                cmap='rainbow',
                edgecolor='k')
    plt.title('Num clusters: {}'.format(n))
    plt.show()
```

```python
#-- testing inertia
k = range(1,8,1)

inertias = []


for i in k:
    kmeans = KMeans(n_clusters=i)

    model = kmeans.fit(X_scaled)

    inertias.append(model.inertia_)
    
plt.plot(k, inertias, '-ob')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.grid()
plt.show()
```

```python
#-- plotting graph with 3 clusters
estimator = KMeans(n_clusters=3)
estimator.fit(X_scaled)
labels = estimator.labels_

plt.scatter(x=X_scaled[:,0],
            y=X_scaled[:,1],
            c=labels.astype(np.float),
            cmap='rainbow',
            edgecolor='k')
plt.title('Num clusters: {}'.format(3))
plt.show()
```
